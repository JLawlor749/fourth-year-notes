# Segmentation 2 - Masking

#### Doing masks to seperate ROIs.

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

I = cv2.imread("orange.png")

I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
Original = I.copy()

plt.imshow(I)
plt.axis('off')
```

```python
V = I[:,:,2]

T = np.mean(V)

T, B = cv2.threshold(V, thresh = T, maxval = 255, type = cv2.THRESH_BINARY)

B = cv2.cvtColor(B, cv2.COLOR_GRAY2RGB)

plt.imshow(B)
plt.axis('off')
```

```python
BN = cv2.bitwise_not(B)

ROI = cv2.bitwise_and(I, BN)

plt.show()
plt.imshow(ROI)
plt.axis('off')
```
