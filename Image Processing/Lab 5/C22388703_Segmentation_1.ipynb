# Segmentation 1

#### Seperating Googly and his friend from the background.

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

I = cv2.imread("Googly.jpg")

I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
Original = I.copy()

plt.imshow(I)

I = cv2.cvtColor(I, cv2.COLOR_RGB2HSV)

V = I[:,:,2]

T = np.mean(V) + (np.std(V))

print(np.mean(V), np.std(V))

T, B = cv2.threshold(V, thresh = T, maxval = 255, type = cv2.THRESH_BINARY)

plt.show()
plt.imshow(B)

B = cv2.cvtColor(B, cv2.COLOR_GRAY2RGB)

maskedImg = cv2.bitwise_and(B, Original)
plt.show()
plt.imshow(maskedImg)
```

#### Testing multiple thresholds to see which is best.

#### It seems like 40 is the best to add to the mean to get the most balanced result.

```python
plt.imshow(Original)

for i in range(0, 20):
    testImg = Original

    I = cv2.cvtColor(testImg, cv2.COLOR_RGB2HSV)

    V = I[:,:,2]

    T = np.mean(V) + i*5

    T, B = cv2.threshold(V, thresh = T, maxval = 255, type = cv2.THRESH_BINARY)

    B = cv2.cvtColor(B, cv2.COLOR_GRAY2RGB)

    maskedImg = cv2.bitwise_and(B, Original)
    print("Added {} to the mean.".format(i*5))
    plt.show()
    plt.imshow(B)
    
```

#### Trying it on the Y in YUV instead.

```python
I = Original

plt.imshow(I)

I = cv2.cvtColor(I, cv2.COLOR_RGB2YUV)

Y = I[:,:,0]

T = np.mean(Y) + (np.std(Y))

print(np.mean(Y), np.std(Y))

T, B = cv2.threshold(Y, thresh = T, maxval = 255, type = cv2.THRESH_BINARY)

plt.show()
plt.imshow(B)

B = cv2.cvtColor(B, cv2.COLOR_GRAY2RGB)

maskedImg = cv2.bitwise_and(B, Original)
plt.show()
plt.imshow(maskedImg)
```
