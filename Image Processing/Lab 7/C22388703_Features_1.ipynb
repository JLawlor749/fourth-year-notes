# Features
### Contours and edges detection.

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image
import math

I = cv2.imread("orange.png")

I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
Original = I.copy()

plt.imshow(I)
plt.axis('off')
```

```python
G = cv2.cvtColor(I, cv2.COLOR_RGB2GRAY)

Ix = cv2.Sobel(G,ddepth=cv2.CV_64F,dx=1,dy=0)
Iy = cv2.Sobel(G,ddepth=cv2.CV_64F,dx=0,dy=1)

M = np.sqrt((Ix * Ix) + (Iy * Iy))

plt.imshow(M, cmap="gray")
plt.axis('off')

E = cv2.Canny(I,threshold1=255,threshold2=0)

plt.show()
plt.imshow(E, cmap="gray")
plt.axis('off')
```

```python
mag_on_edge = cv2.bitwise_and(M, M, mask=E)
plt.imshow(mag_on_edge, cmap="gray")
plt.axis('off')
```
