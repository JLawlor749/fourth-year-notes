# Image Maths

### Adding, multiplying images.

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

water = cv2.imread('water.jpg')
orange = cv2.imread('orange.png')

water = cv2.cvtColor(water, cv2.COLOR_BGR2RGB)
orange = cv2.cvtColor(orange, cv2.COLOR_BGR2RGB)

water2 = water.copy()
orange2 = orange.copy()

water = cv2.multiply(water, 0.4, scale = 1)
orange = cv2.multiply(orange, 0.6, scale = 1)

newImg = cv2.add(water, orange)
weightedImg = cv2.addWeighted(water2, 0.4, orange2, 0.6, 1 )



plt.imshow(water)
plt.show()
plt.imshow(orange)
plt.show()
plt.imshow(newImg)
plt.show()
plt.imshow(weightedImg)





```
