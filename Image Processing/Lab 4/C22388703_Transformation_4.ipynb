# Kernels

### Sharpening Image

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

celeste = cv2.imread('celeste.png')
celeste = cv2.cvtColor(celeste, cv2.COLOR_BGR2RGB)

h, w, d = celeste.shape
#celeste = cv2.resize(celeste, dsize=(3*w, 3*h))

plt.imshow(celeste)

k = np.array([[-1,-2,-1], [-2,13,-2], [-1,-2,-1]], dtype=float)

sharp = cv2.filter2D(celeste,ddepth=-1,kernel=k)

plt.show()
plt.imshow(sharp)

subtracted = cv2.subtract(sharp,celeste)

plt.show()
plt.imshow(subtracted)
```
