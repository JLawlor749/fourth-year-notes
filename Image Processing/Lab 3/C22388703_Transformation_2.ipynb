# Transformation

### Cropping Images

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image
import math

I = cv2.imread('superman.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

Original = I.copy()

h, w, c = I.shape

qh = int(h/2)
qw = int(w/2)

Q = I[qh:h, qw:w]

plt.imshow(Q)



```

### Scaling and Rotating

```python
ch, cw, cc = Q.shape

M = cv2.getRotationMatrix2D(center=(cw/2,ch/2), angle=45, scale=0.50)
R = cv2.warpAffine(Q, M = M, dsize=(cw,ch))

plt.imshow(R)
```
