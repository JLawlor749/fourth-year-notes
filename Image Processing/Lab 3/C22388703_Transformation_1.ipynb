# Transformations

### Plotting Histograms

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

I = cv2.imread('wartime.jpg')
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)

Original = I.copy()

Values = I.ravel()

plt.subplot(1, 2, 1)
plt.hist(Values,bins=256,range=[0,256])

plt.subplot(1, 2, 2)
plt.imshow(I, cmap="gray")
```

### Equalizing Histograms

```python
E = cv2.equalizeHist(I)

Values = E.ravel()

plt.subplot(1, 2, 1)
plt.hist(Values,bins=256,range=[0,256])

plt.subplot(1, 2, 2)
plt.imshow(E, cmap="gray")
```
