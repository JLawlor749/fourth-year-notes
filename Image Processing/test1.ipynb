```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

imageIn = cv2.imread("Group Project/ReceiptReader/testdata/bended1.png")

imageRGB = cv2.cvtColor(imageIn, cv2.COLOR_BGR2RGB)

imageIn = cv2.cvtColor(imageIn, cv2.COLOR_BGR2GRAY)
imageOriginal = imageIn.copy()

plt.imshow(imageRGB)
plt.axis("off")

imgThresh = np.mean(imageIn) - np.std(imageIn)

imgThresh, output = cv2.threshold(imageIn, thresh = imgThresh, maxval = 255, type = cv2.THRESH_BINARY)

plt.show()
plt.imshow(output, cmap="gray")
plt.axis("off")
```
