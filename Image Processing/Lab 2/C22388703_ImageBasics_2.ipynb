### Drawing on Images

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

I = cv2.imread('dleju37fj6761.png')

Original = I.copy()

cv2.line(img = I, pt1 = (0,0), pt2 =
(604,800), color = (255,255,255), thickness = 5)

I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
plt.imshow(I)
plt.axis('off')

```

### Drawing a Circle

```python
cv2.circle(img = I, center = (302,260), radius =
125, color = (255,0,0), thickness = 10)

plt.imshow(I)
plt.axis('off')
```

### Drawing a Rectangle

```python
cv2.rectangle(img = I, pt1 = (200,350), pt2 =
(404,700), color = (0,255,255), thickness = 10)

print(I.shape)

plt.imshow(I)
plt.axis('off')
```

### Capturing user input to draw on an image.

```python
def draw(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img = I, center = (x,y),
        radius = 5, color = (255,255,255),
        thickness = -1)
        cv2.imshow("image", I)


I = Original.copy()

cv2.imshow("image", I)
cv2.namedWindow("image")
cv2.setMouseCallback("image", draw)
cv2.waitKey(0)
```

### Capturing user input to convert a rectangle to YUV.

```python
def rectToYUV(event,x,y,flags,param):
    
    if event == cv2.EVENT_LBUTTONDOWN:

        if x-100 < 0:
            x = 100

        if y-100 < 0:
            y = 100

        segment = I[y-100:y+100, x-100:x+100]
        segment = cv2.cvtColor(segment, cv2.COLOR_BGR2YUV)

        cv2.rectangle(img = I, pt1 = (x-100,y-100), pt2 =
        (x+100,y+100), color = (0,0,255), thickness = 5)

        I[y-100:y+100, x-100:x+100] = segment

        cv2.imshow("image", I)


I = Original.copy()

cv2.imshow("image", I)
cv2.namedWindow("image")
cv2.setMouseCallback("image", rectToYUV)
cv2.waitKey(0)
```
