### Importing an image.

```python
import cv2 as cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as image

I = cv2.imread('All_Might.webp')
```

### Showing an image.

```python
RGB = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
plt.imshow(RGB)
plt.axis('off')  # Hide the axis labels
```

### Changing colour space to YUV.

```python
Y, U, V = cv2.split(cv2.cvtColor(RGB, cv2.COLOR_RGB2YUV))
plt.imshow(Y, cmap="gray")
plt.axis('off')  # Hide the axis labels
```

### Convert the colour space to greyscale.

```python
G = cv2.cvtColor(RGB, cv2.COLOR_RGB2GRAY)
plt.imshow(G, cmap="gray")
plt.axis('off')  # Hide the axis labels
```

### Showing luminance alongside intensity.

```python
luminance, U, V = cv2.split(cv2.cvtColor(RGB, cv2.COLOR_RGB2YUV))
plt.subplot(1, 2, 1)
plt.imshow(luminance, cmap="gray")
plt.axis('off')  # Hide the axis labels
plt.title("Luminance")

H, saturation, intensity = cv2.split(cv2.cvtColor(RGB, cv2.COLOR_RGB2HSV))
plt.subplot(1, 2, 2)
plt.imshow(intensity, cmap="gray")
plt.axis('off')  # Hide the axis labels
plt.title("Intensity")
```
